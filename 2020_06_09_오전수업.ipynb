```
#시간과 관련된 기능가져오기
import time
 
#변수 선언
number = 0

target_tick = time.time() + 5
while time.time() < target_tick:
  number += 1

print("5초동안 {}번 반복했습니다.".format(number))
```

break 키워드

반복문 벗어날떄 사용하는 키워드

```
#변수 선언
i = 0

#무한반복

while True:
  print("{}번째 반복문입니다.".format(i))
  i = i + 1 
  input_text = input(">종료하시겠습니까?(y):")
  if input_text in ["y", "Y"]:
    print("반복을 종료합니다.")
    break
```

continue 키워드

현재 반복을 생략하고 다음 반복으로 넘어감

```
#변수 선언
numbers = [5, 15, 6, 20, 7, 25]

#반복
for number in numbers:
  #number가 10보다 작으면 다음 반복으로 넘어간다
  if number < 10:
    continue
  print(number)
```

reversed()함수

리스트에서 요소 뒤집기

```
list_a = [1, 2, 3, 4, 5]
list_reversed = reversed(list_a)

print("reversed([1, 2, 3, 4, 5]):", list_reversed)
print("list(reversed([1, 2, 3, 4, 5])):", list(list_reversed))
print()

print("#reversed()함수와 반복문")
print("for i in versed([1, 2, 3, 4, 5]):")

for i in reversed(list_a):
  print("-", i)
```

```
#temp는 반복적으로 사용안됨
temp = reversed([1, 2, 3, 4, 5])

for i in temp:
  print(i)
```

```
#반복적으로 사용가능
numbers = [1, 2, 3, 4, 5]
for i in reversed(numbers):
  print(i)
```

enumerate()함수

리스트 요소 반복할때 현재 인덱스가 몇번째인지 확인

example_list = ["요소A", "요소B", "요소C"]

```
example_list = ["요소A", "요소B", "요소C"]

print("#단순 출력")
print(example_list)
print()

#enumerate() 함수를 적용해출력
print("#enumerate() 함수를 적용해출력")
print(enumerate(example_list))
print()

print(list(enumerate(example_list)))

# for반복문
for i, value in enumerate(example_list):
  print("{}번째 요소는 {}입니다.".format(i + 1, value))
```

딕셔너리와 itmes()함수 함께 사용

```
example_dictionary = {
    "키A": "값A",
    "키B": "값B",
    "키C": "값C",
}

print("items():", example_dictionary.items())

for key, element in example_dictionary.items():
  print("dictionary[{}] = {}".format(key, element))
```
